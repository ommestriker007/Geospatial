```python
import ee
import folium
import geemap
```

```python
ee.Authenticate()
ee.Initialize(project = "neon-feat-313910")
```

```python
#Solapur district coordinates

aoi = ee.Geometry.Rectangle([74.25, 17.10, 76.00, 18.40])
```

```python
# Landsat 8 surface reflectance image
landsat = ee.ImageCollection("LANDSAT/LC08/C02/T1_L2") \
    .filterBounds(aoi) \
    .filterDate('2023-01-01', '2024-12-31') \
    .filter(ee.Filter.lt('CLOUD_COVER', 20))

landsat_mosaic = landsat.median()
```

```python
# Extracting thermal band and apply scaling factors
thermal = landsat_mosaic.select('ST_B10').multiply(0.00341802).add(149.0).subtract(273.15)
```

```python
Map = geemap.Map(center=[18.5, 73.9], zoom=10)

# Adding LST layer
Map.addLayer(thermal.clip(aoi),
             {'min': 20, 'max': 45, 'palette': ['blue', 'cyan', 'green', 'yellow', 'red']},
             'Land Surface Temperature (Â°C)')

Map.centerObject(aoi, 9)
Map
```
